### define the calculator

```python
def calci(x,y, intuition):
    if intuition == 'add':
        return x+y
    elif intuition == 'sub':
        return x-y
    elif intuition == 'multiply':
        return x*y
    elif intuition == 'divide':
        return x/y
    else:
        return 'invalid Intuition'
```

```python
calci(25,25, 'multiply')
```

```python
calci(25,25, 'add')
```

### count the vowels in the given string

```python
def vowel(s):
 vowels = 'AEIOUaeiou'
 count = sum(1 for char in s if char in vowels)
 return count
```

### filter() function ----- to filter out the values we want

```python
# filter the positive numbers from the mixed data
mylist = [78,5,-90,44,-0,-5,5.5,4,-2]
flt = list(filter(lambda x: x >= 0 , mylist))
```

```python
flt
```

### map() function ---- provides the output after executing the function to each item in the sequence


```python
# create a map function to cube of all elements
mylist = [5,6,8,9,7,10]
cube = list(map(lambda x: x**3, mylist))
```

```python
cube
```

```python
# convert name of all animals to uppercase
anml = ['parrot','tiger','elephant','bear']
anml = list(map(lambda x: x.upper(), anml))
anml
```
