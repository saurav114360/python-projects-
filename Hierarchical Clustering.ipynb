### Hierarchical Clustering

#### Import the Data Set

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Gaurav\Downloads\customers.csv")
```

```python
df.head()
```

### Note- To apply Hier...Clustering we will take Annual income and Spending score columns


```python
X = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
X.head()
```

```python
X.shape
```

### Plotting the scatter chart to see the distribution


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

### Plot the dendrogram , which predict the optimal number of clusters


```python
import scipy.cluster.hierarchy as hc
```

```python
plt.title('Customers dendrogram')
dendo = hc.dendrogram(hc.linkage(X, method = 'ward'))
```

### Now we have selected the optimal number of clusters -- 5

##### we will plot the graph to show these clusters

```python
from sklearn.cluster import AgglomerativeClustering as AC
```

```python
cluster = AC(n_clusters = 5, linkage='ward')
```

```python
labels = cluster.fit(X)
```

```python
# check the labels or clusters values
labels
```

```python
labels
```

```python
labels = cluster.fit_predict(X)
labels
```

### Now we will plot the graph

```python
X.columns
```

```python
plt.figure(figsize = (10, 7))
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'],
 c = labels)
plt.show()
```

```python
plt.figure(figsize = (10, 7))
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'],
 c = labels, cmap = 'magma')
plt.show()
```
