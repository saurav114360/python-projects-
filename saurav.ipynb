```python
import pandas as pd

# Function to add a new column
def add_salary(df, base_salary):
    df['Salary'] = base_salary + (df['Age'] * 100)
    return df

# Creating a DataFrame
data = {'Name':['Saurav','Garv','Sanskar','Sanskar'],
        'Age': [25,30,35,40]}
df = pd.DataFrame(data)

# Applying the function
df_with_salary = add_salary(df,20000)
print(df_with_salary)


```

```python
import pandas as pd 

# Function to calculate years until retirement (assume retirement age is 65)
def year_until_retirement(age):
    return 65 - age

# creating a DataFrame
data = {'Name':['Saurav','Garv','Sanskar','Sanskar'],
        'Age': [25,30,35,40]}
df = pd.DataFrame(data)

# Adding a new column with the calculate values
df['year until retirement'] = df['Age'].apply(year_until_retirement)
print(df)
```

```python
#Column_Wise Arithmetic operations

df=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})

df['Column_sum']=df.apply(lambda column: column['A']+column['B'],axis=1)
df['Column_product']=df.apply(lambda column: column['A']*column['B'],axis=1)
print(df)
```

```python
#Raw_Wise Arithmetic operations

df=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})

df['Raw_sum']=df.apply(lambda row: row['A']+row['B'],axis=1)
df['Raw_product']=df.apply(lambda row: row['A']*row['B'],axis=1)
print(df)
```

```python
df['column_sum']=df.apply(np.sum,axis=1)
print(df)
```
