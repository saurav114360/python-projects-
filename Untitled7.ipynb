### we want to group spending score with respect to age/ we need age and spending score

```python
# define the data
x=df[['Age','Spending Score (1-100)']]
```

### visualize your unclustered data with scatter plot

```python
# with maplotlib
plt.scatter(x['Age'],x['Spending Score (1-100)'])
plt.xlabel('Age of the Employee')
plt.ylabel('Spending score')
plt.show()
```

### loading the k-means clustering algo..

```python
from sklearn.cluster import KMeans
```

### finding the optimal number of cluster

```python
error = []

for i in range(1,11):
    model1 = KMeans(n_clusters = i)
    model1.fit(x)
    error.append(model1.inertia_)
```
