### simple linear regression model

```python
# import the libraries 
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv(r"C:\Users\Gaurav\Downloads\data.csv")
df.head()
```

```python
# check the statistical information of dataset
df.describe()
```

### split the dataset into x (input) and y (output)

```python
# input
X = df[['YearsExperience']]
```

```python
X.head()
```

```python
X.shape
```

```python
Y=df[['Salary']]
```

```python
Y.head()
```

### split the data into training and testing phase

```python
# import the train_test_split
from sklearn.model_selection import train_test_split
```

```python
from sklearn import model_selection
```

```python
print(dir(model_selection))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8, random_state = 42)
```

```python
print('shape of X_train:', X_train.shape)
print('shape of X_test:', X_test.shape)
print('shape of Y_train:', Y_train.shape)
print('shape of Y_test:', Y_test.shape)
```

```python
X_train[:5]
```

### building the linear regression model for salary prediction

```python
# import the required algorithm
from sklearn.linear_model import LinearRegression
```

```python
# now we create the object for Linear Regession
LinearModel = LinearRegression()
```

```python
# its time to train our model
# for train the model we will be using the "fit()" function

LinearModel.fit(X_train, Y_train)
```

### now our model has learnt every thing , now its time to predict the values

```python
Y_pred = LinearModel.predict(X_test)
```

```python
Y_pred # prediction salary from our model
```

```python
Y_test
```

```python
Y_pred[:5]
```

### check the model performance

```python
from sklearn.metrics import r2_score
```

```python
print (r2_score(Y_test,Y_pred))
```

```python
print (r2_score(Y_test,Y_pred)*100)
```
