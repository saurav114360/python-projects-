### MULTIPLE LINER REGRESSION MODEL


```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# to ignore the unwanted future warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\Gaurav\Downloads\telecom_data.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# splitting data x (input) and y (output) features
```

```python
# input 
X = df.drop(columns = ['Average_Data_Usage_Per_Month'])
```

```python
X.head()
```

```python
X.shape
```

```python
# define the output column 
y = df[['Average_Data_Usage_Per_Month']]
```

```python
y.head()
```

```python
# now we have to split our data into 80:20 ratio test split 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 42)
```

```python
# shape of the splitted dataset
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# data normalization ----standard scaler
```

```python
# import the model
from sklearn.preprocessing import StandardScaler
```

```python
# object for model 
sc = StandardScaler()
```

```python
sc.fit(X_train)
X_train_sc= sc.transform(X_train)
X_test_sc= sc.transform(X_test) 
```

```python
X_train_sc
```

```python
pd.DataFrame(X_train_sc)
```

### loading the model

```python
from sklearn.linear_model import LinearRegression
```

```python
# create an object for model 
model1 = LinearRegression()
```

```python
# fitting or train the data 
model1.fit(X_train_sc, y_train)
```

### now our model has learnt evrtything and ready to prediction

```python
y_pred = model1.predict(X_test_sc)
```

```python
y_pred
```

```python
# compare the predict 
y_test
```

### CHECKING THE PERFORMANCE OF THE MODEL

```python
from sklearn.metrics import r2_score
```

```python
#accuracy
print(r2_score(y_test,y_pred))
```

### OUR MODEL HAS GIVEN 100% ACCURACY
