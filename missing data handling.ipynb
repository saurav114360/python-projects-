### filing missing data with median for Age column

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

```python
# filling the data for salary?assigment
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
```

```python
df.isnull().sum()
```

```python
# working on categorical type of data for department
df['Department'].mode()
```
